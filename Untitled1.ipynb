```python
df_tetracyc = df[df["class"] == "TETRACYCLINE"]
df_tetracyc
```

```python
df1 = pd.DataFrame(columns=("subject", "predicate", "object"))
df1["subject"] = df_tetracyc["subclass"].unique()
df1["predicate"] = "isSubclassOf"
df1["object"] = "TETRACYCLINE"
df1 = df1[df1["subject"] != df1["object"]] # drop self loops
df1.reset_index(drop=True, inplace=True)
```

```python
df2 = pd.DataFrame(columns=("subject", "predicate", "object"))
df2["subject"] = df_tetracyc["gene_family"]
df2["object"] = df_tetracyc["subclass"]
df2["predicate"] = "isMadeBy"

df2.reset_index(drop=True, inplace=True)

df = pd.concat((df1, df2)) # .drop_duplicates("subject")))

df
```

```python
import networkx as nx
G=nx.from_pandas_edgelist(df, "object", "subject", edge_attr=True)
```

```python

plt.figure(figsize=(12, 12))

pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_color="red", edge_cmap=plt.cm.Blues, pos=pos)
plt.show()

```

```python
df_tigecyc = df_tetracyc[df_tetracyc["subclass"] == "TIGECYCLINE"]

df1 = pd.DataFrame(columns=["subject", "predicate", "object"])
df1["object"] = df_tigecyc["class"]
df1["subject"] = df_tigecyc["gene_family"]
df1["predicate"] = "belongsTo"
df1.drop_duplicates(inplace=True)
```

```python
df2 = pd.DataFrame(columns=["subject", "predicate", "object"])
df2["subject"] = df_tigecyc["allele"] 
df2["object"] = df_tigecyc["gene_family"] 
df2["predicate"] = "belongsTo"
df2.dropna(inplace=True)
```

```python
df = pd.concat((df1, df2), ignore_index=True)
df
```

```python
G=nx.from_pandas_edgelist(df, "object", "subject", edge_attr=True)
plt.figure(figsize=(12, 12))

pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_color="red", edge_cmap=plt.cm.Blues, pos=pos)
plt.show()
```

```python
df3 = pd.DataFrame(columns=["subject", "predicate", "object"])
df3["subject"] = df_tigecyc["allele"]
df3["object"] = df_tigecyc["product_name"] 
df3["predicate"] = "produces"
df3 # what to do with Nan?
```

```python
df = pd.concat((df, df3))
df.dropna(inplace=True)
df
```
